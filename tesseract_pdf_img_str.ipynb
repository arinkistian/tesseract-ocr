```python
@title Default code

import pdf2image #convert pdf to image
from PIL import Image #library PIL; work with image
import pytesseract
from google.colab import files #upload file

# conver pdf to image
def pdf_to_img(pdf_file):
    return pdf2image.convert_from_path(pdf_file)

# OCR pada gambar.
def ocr_core(file):
    text = pytesseract.image_to_string(file)
    return text

# extract text from first page
def extract_text_from_first_page(pdf_file):
    images = pdf_to_img(pdf_file)
    first_page_img = images[0]
    text_first_page = ocr_core(first_page_img)
    return text_first_page

# extract title
def extract_judul(text):
    start = text.find("Judul Buku")
    end = text.find("\n", start)
    judul = text[start:end].replace("Judul Buku", "").strip()
    return judul

# extract isbn
def extract_isbn(text):
    start = text.find("ISBN")
    end = text.find("\n", start)
    isbn = text[start:end].replace("ISBN", "").strip()
    return isbn

# extraxt author info
def extract_penulis(text):
    start = text.find("Penulis")
    end = text.find("\n", start)
    penulis = text[start:end].replace("Penulis", "").strip()
    return penulis

# extract pdf info
def print_info(pdf_file):
    # Extract and print information for the first page
    text_first_page = extract_text_from_first_page(pdf_file)
    judul = extract_judul(text_first_page)
    isbn = extract_isbn(text_first_page)
    penulis = extract_penulis(text_first_page)

    print("Informasi (Halaman 1):")
    print("Judul Buku:", judul)
    print("ISBN:", isbn)
    print("Penulis:", penulis)

    # Extract and print information for all pages
    images = pdf_to_img(pdf_file)
    all_pages_text = ""
    for pg, img in enumerate(images):
        all_pages_text += ocr_core(img)

# upload pdf file
def select_pdf_file():
    uploaded = files.upload()
    for file_name in uploaded.keys():
        return file_name

if __name__ == "__main__":
    # choose PDF file
    pdf_file = select_pdf_file()
    if pdf_file:
        print_info(pdf_file)
    else:
        print("No PDF file selected.")
```

```python
from PIL import Image
import pytesseract
from google.colab import files
import os

def ocr_core(image):
    text = pytesseract.image_to_string(image)
    return text

def extract_isbn(text):
    start = text.find("ISBN")
    end = text.find("\n", start)
    isbn = text[start:end].replace("ISBN", "").strip()
    return isbn

def extract_penulis(text):
    start = text.find("Penulis")
    end = text.find("\n", start)
    penulis = text[start:end].replace("Penulis", "").strip()
    return penulis

def collect_ground_truth_data():
    ground_truth_data = {}

    print("Please upload image files. Type 'done' when finished.")
    while True:
        file_name = select_image_file()
        if file_name:
            with open(file_name, "rb") as file:
                text_data = file.read()
            ground_truth_data[file_name] = text_data
            print("File {} uploaded and text data saved.".format(file_name))
        else:
            print("No file uploaded.")

        stop_collecting = input("Do you want to stop collecting data? (yes/no): ").lower()
        if stop_collecting == "yes":
            break

    return ground_truth_data

def calculate_word_error_rate(actual_text, ocr_text):
    actual_words = actual_text.lower().split()
    ocr_words = ocr_text.lower().split()

    total_words = len(actual_words)
    word_errors = sum(1 for word in ocr_words if word not in actual_words)

    if total_words == 0:
        word_error_rate = 0
    else:
        word_error_rate = word_errors / total_words

    return word_error_rate

def calculate_recall_precision(ground_truth_data, ocr_data):
    total_ground_truth_words = 0
    total_correctly_detected_words = 0
    total_detected_words = 0

    for file_name, ground_truth_text in ground_truth_data.items():
        ocr_text = ocr_data.get(file_name, "")

        ground_truth_words = set(ground_truth_text.lower().split())
        ocr_words = set(ocr_text.lower().split())

        correctly_detected_words = ground_truth_words.intersection(ocr_words)
        total_correctly_detected_words += len(correctly_detected_words)

        total_ground_truth_words += len(ground_truth_words)
        total_detected_words += len(ocr_words)

    if total_ground_truth_words == 0:
        # Handle the case when ground_truth_data is empty to avoid division by zero
        recall = 0
    else:
        recall = total_correctly_detected_words / total_ground_truth_words

    if total_detected_words == 0:
        # Handle the case when there are no detected words to avoid division by zero
        precision = 0
    else:
        precision = total_correctly_detected_words / total_detected_words

    return recall, precision

if __name__ == "__main__":
    ground_truth_data = collect_ground_truth_data()

    ocr_results = {}
    for image_file, actual_text in ground_truth_data.items():
        print("Processing image:", image_file)

        image = Image.open(image_file)

        text_data_buku = ocr_core(image)
        isbn = extract_isbn(text_data_buku)
        penulis = extract_penulis(text_data_buku)

        print("ISBN:", isbn)
        print("Penulis:", penulis)

        ocr_results[image_file] = text_data_buku

    recall, precision = calculate_recall_precision(ground_truth_data, ocr_results)
    print("Recall: {:.2f}".format(recall))
    print("Precision: {:.2f}".format(precision))
```
